```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the cleaned dataset (ensure 'cleaned_iris.csv' is uploaded or use the original dataset)
iris_data = pd.read_csv('/content/sample_data/cleaned_iris.csv')

# Display the first few rows
iris_data.head()
```

```python
# Bar plot of average petal length for each species
plt.figure(figsize=(8, 6))
sns.barplot(data=iris_data, x='species', y='petal_length', palette='viridis')
plt.title('Average Petal Length for Each Species', fontsize=16)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Average Petal Length', fontsize=12)
plt.show()
```

```python
# Line chart of petal length vs. sepal length for each species
plt.figure(figsize=(8, 6))
for species in iris_data['species'].unique():
    subset = iris_data[iris_data['species'] == species]
    plt.plot(subset['sepal_length'], subset['petal_length'], marker='o', label=species)

plt.title('Petal Length vs. Sepal Length by Species', fontsize=16)
plt.xlabel('Sepal Length', fontsize=12)
plt.ylabel('Petal Length', fontsize=12)
plt.legend(title='Species')
plt.grid()
plt.show()
```

```python
# Scatter plot of sepal width vs. petal width
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris_data, x='sepal_width', y='petal_width', hue='species', palette='cool')
plt.title('Sepal Width vs. Petal Width by Species', fontsize=16)
plt.xlabel('Sepal Width', fontsize=12)
plt.ylabel('Petal Width', fontsize=12)
plt.legend(title='Species')
plt.show()
```
